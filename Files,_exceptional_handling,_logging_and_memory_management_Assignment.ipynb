23. The challenges associated with memory management in Python are as follows:-
    
    High Memory Usage: Python objects often have a higher memory footprint compared to objects in lower-level languages like C or C++. This can lead to applications consuming more RAM than expected, especially when dealing with large datasets or long-running processes.

    Garbage Collection Overhead: While Python's automatic garbage collection simplifies development, it can introduce performance overhead. Frequent or complex garbage collection cycles can slow down application execution, particularly when memory usage is high or reference cycles are present.

    Circular References: Python's primary garbage collection mechanism, reference counting, cannot detect and collect objects involved in circular references (where objects directly or indirectly refer to each other in a closed loop). This can lead to memory leaks if not addressed by the generational garbage collector.

    Fragmentation: Memory fragmentation can occur when memory is allocated and deallocated in non-contiguous blocks, leaving small, unusable gaps. This can reduce the efficiency of memory allocation and potentially prevent the allocation of larger contiguous blocks, even if sufficient total memory is available.

    Global Interpreter Lock (GIL): In CPython, the GIL ensures that only one thread can execute Python bytecode at a time. While this simplifies memory management in multi-threaded environments by preventing race conditions on shared memory, it can limit the benefits of multi-threading for CPU-bound tasks, as true parallel execution of Python code is not achieved.

    Limited Manual Control: Python's automatic memory management reduces the need for manual memory handling but also offers less fine-grained control compared to languages with explicit memory management. This can make it challenging to optimize memory usage in highly performance-critical scenarios without resorting to lower-level extensions or careful design choices.

3. If a file doesn't exist when I try to open it in read mode ("r"), Python raises a FileNotFoundError. I can handle this situation gracefully using a try...except block.

```python
try:
    # Attempt to open the file
    with open("example.txt", "r") as file:
        # Read and print each line
        for line in file:
            print(line.strip())

except FileNotFoundError:
    # Handle the case where the file doesn't exist
    print("Error: The file was not found.")

except Exception as e:
    # Handle any other unexpected errors
    print(f"An unexpected error occurred: {e}")
```

```python
# Open the file and read all lines into a list
with open("example.txt", "r") as file:
    lines = file.readlines()

print(lines)
```

Using a Loop (More Memory Efficient)

```python
lines = []

with open("example.txt", "r") as file:
    for line in file:
        lines.append(line.strip())  # strip() removes newline characters

print(lines)
```

```python
with open("example.txt", "r") as file:
    lines = [line.strip() for line in file]

print(lines)
```

18. I can implement a basic logging setup with file rotation in Python using the RotatingFileHandler from the logging.handlers module. This allows the log file to roll over when it reaches a certain size (eg 1 MB).

```python
import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.INFO)  # Log level

# Create a RotatingFileHandler
handler = RotatingFileHandler(
    "app.log",      # Log file name
    maxBytes=1_000_000,  # 1 MB
    backupCount=3        # Keep up to 3 backup files
)

# Create a log message format
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example logging
for i in range(10000):
    logger.info(f"This is log message number {i}")

print("Logging complete with rotation setup.")
```

22. I can check if a file is empty before reading using either "os.path.getsize"() or the "pathlib module". This prevents reading an empty file and helps handle it gracefully.

Using os.path.getsize()

```python
import os

file_path = "example.txt"

if os.path.exists(file_path):
    if os.path.getsize(file_path) > 0:
        with open(file_path, "r") as file:
            content = file.read()
            print(content)
    else:
        print("The file is empty.")
else:
    print("Error: The file does not exist.")
```

Using pathlib.Path

```python
from pathlib import Path

file_path = Path("example.txt")

if file_path.exists():
    if file_path.stat().st_size > 0:
        with open(file_path, "r") as file:
            print(file.read())
    else:
        print("The file is empty.")
else:
    print("Error: The file does not exist.")
```
